# Soal 1, Akun Bank

```python
class bank:
    def __init__(self, nomor_akun, pemilik_akun, saldo=0):
        self.nomor_akun = nomor_akun
        self.pemilik_akun = pemilik_akun
        self.saldo = saldo

    def deposit(self, jumlah):
        if jumlah > 0:
            self.saldo += jumlah
            print(f"Deposit {jumlah} ke akun{self.nomor_akun}. saldo sekarang: {self.saldo}")
        else:
            print("invalid jumlah deposit ")

    def penarikan(self, jumlah):
        if 0 < jumlah <= self.saldo:
            self.saldo -= jumlah
            print(f"penarikan sebanyak {jumlah} dari akun {self.nomor_akun} saldo sekarang: {self.saldo}")
        else:
            print("penarikan invalid.")

    def info_akun(self):
        print(f"nomor_akun: {self.nomor_akun}")
        print(f"pemilik_akun: {self.pemilik_akun}")
        print(f"saldo: {self.saldo}")

username = input('Masukkan Username Anda: ')
password = input('Masukkan Password Anda: ')

if username == 'akusayangkamu' and password == '1234':
    print('Login Sukses!')
    akun1 = bank("54321", "kelompok ku", 5000)
    akun1.info_akun()
    akun1.deposit(500)
    akun1.penarikan(300)
elif username != 'akusayangkamu' and password != '1234':
    print('Login Gagal (username dan password salah)')
elif username != 'akusayangkamu':
    print('Login Gagal (username salah)')
elif password != '1234':
    print('Login Gagal (password salah)')
```

# Soal 2,Data Karyawan

```python
class Karyawan:
    def __init__(self, id_karyawan, nama, gaji_bulanan):
        self.id_karyawan = id_karyawan
        self.nama = nama
        self.gaji_bulanan = gaji_bulanan

    def hitung_gaji_tahunan(self):
        return self.gaji_bulanan * 12

    def tampilkan_detail_karyawan(self):
        print(f"ID Karyawan: {self.id_karyawan}")
        print(f"Nama: {self.nama}")
        print(f"Gaji Bulanan: Rp {self.gaji_bulanan:,}")
        print(f"Gaji Tahunan: Rp {self.hitung_gaji_tahunan():,}")

karyawan1 = Karyawan(1, "Lulu", 3000000)
karyawan2 = Karyawan(2, "Lala", 8000000)

karyawan1.tampilkan_detail_karyawan()
print("\n")
karyawan2.tampilkan_detail_karyawan()
```

# Soal 3,Barang Inventaris

```python
class Barang:
    def __init__(self, id_barang, nama, jumlah, harga):
        self.id_barang = id_barang
        self.nama = nama
        self.jumlah = jumlah
        self.harga = harga

    def tambah_produk(self, jumlah_tambahan):
        self.jumlah += jumlah_tambahan
        print(f"Jumlah produk {self.nama} ditambahkan sebanyak {jumlah_tambahan}.")

    def memperbarui_jumlah(self, jumlah_awal, jumlah_tambahan):
        self.jumlah = jumlah_awal + jumlah_tambahan
        print(f"Jumlah produk {self.nama} diperbarui menjadi {self.jumlah}.")

    def hitung_total_harga(self):
        total_harga = self.jumlah * self.harga
        return total_harga

    def tampilkan_detail_inventaris(self):
        print(f"ID Barang: {self.id_barang}")
        print(f"Nama: {self.nama}")
        print(f"Jumlah: {self.jumlah}")
        print(f"Harga: {self.harga:,}")
        print(f"Total Harga: {self.hitung_total_harga():,}")

barang1 = Barang(id_barang="004pdr", nama="Tas", jumlah=3, harga=4000000)

barang1.tambah_produk(jumlah_tambahan=2)

barang1.memperbarui_jumlah(jumlah_awal=3, jumlah_tambahan=2)

barang1.tampilkan_detail_inventaris()
```
