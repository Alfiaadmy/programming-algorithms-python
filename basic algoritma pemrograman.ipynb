### Soal 1, Perkalian List

Seorang guru matematika memberikan tugas kepada murid-muridnya untuk membuat program sederhana yang dapat menghitung hasil perkalian dari sejumlah bilangan bulat yang dimasukkan. Bantulah murid-murid tersebut dengan membuat program sesuai spesifikasi berikut:

Spesifikasi Program:
- Program menerima masukan berupa sejumlah bilangan bulat.
- Program kemudian menghitung hasil perkalian dari semua bilangan yang dimasukkan.
- Program menggunakan error handling untuk menangani masukan yang tidak valid, seperti huruf, karakter khusus, atau masukan yang bukan bilangan bulat.
- Jika masukan tidak valid, program memberikan pesan kesalahan dan meminta pengguna memasukkan bilangan bulat kembali.
- Hasil perkalian dari bilangan yang valid kemudian ditampilkan sebagai keluaran program.

Catatan:
- Program harus mampu menangani bilangan bulat positif dan negatif.
- Pengguna dapat mengakhiri input dengan memasukkan angka 0.
- Program memberikan pesan kesalahan jika masukan tidak valid dan meminta pengguna untuk memasukkan kembali.

```python
print("Selamat datang di Kalkulator Perkalian Bilangan Bulat!")
print("Masukkan bilangan bulat. Ketik 0 untuk mengakhiri.")
print()

def hitung_perkalian():
    hasil = 1
    while True:
        try:
            bilangan = int(input("Masukkan bilangan : "))
            if bilangan == 0:
                break
            hasil *= bilangan
        
        except ValueError:
            print("Masukkan tidak valid. Harap masukkan bilangan bulat.")
    if hasil != 1:
        print()
        print(f"Terima kasih! Hasil perkalian adalah: {hasil}")
hitung_perkalian()
```

### Soal 2, Data Karyawan

Seorang guru bahasa ingin mengajarkan murid-muridnya untuk membuat program sederhana yang dapat memproses kalimat dan mengeluarkan semua kata yang terdapat dalam kalimat tersebut. Bantulah murid-murid tersebut dengan membuat program sesuai spesifikasi berikut:

Spesifikasi Program:

- Program menerima masukan berupa sebuah kalimat.
- Program kemudian memproses kalimat tersebut untuk mengidentifikasi dan mengekstrak semua kata yang terdapat dalam kalimat.
- Setiap kata yang diekstrak kemudian ditampilkan sebagai keluaran program, tanpa ada duplikasi kata.
- Program menggunakan error handling untuk menangani masukan yang tidak valid, seperti kalimat kosong atau masukan selain teks.
- Output program harus mencerminkan setiap kata hanya sekali, meskipun kata tersebut muncul beberapa kali dalam kalimat.

```python
dipisah=[]
kata=""
spasi=0
while True:
    try:
        kalimat_input=input("Masukkan kalimat anda: ")
        if len(kalimat_input)==0:
            raise ValueError
        else:
            for k in (f"{kalimat_input} "):
                if k!=" ":
                    kata+=k
                if k==" ":
                    if kata in dipisah:
                        kata=""
                        continue
                    dipisah.append(kata)
                    kata=""
            for i in dipisah:
                print(f"-{i}")
            break
    except ValueError:
        print("Inputmu kosong kawan")
```
